```python
from sklearn.cluster import AgglomerativeClustering
```

```python
from sklearn.datasets import make_blobs 
```

```python
x,y = make_blobs(random_state=0, n_samples=1000)
```

```python
x
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x[:,0],x[:,1],'ro')
plt.show()
```

```python
clustering = AgglomerativeClustering(n_clusters=3,linkage='ward',affinity='euclidean')
```

```python
clustering.fit(x)
```

```python
plt.scatter(x[:,0],x[:,1],c=clustering.labels_, marker='o')
plt.show()
```

```python
clustering = AgglomerativeClustering(n_clusters=2,linkage='average',affinity='manhattan')
```

```python
clustering.fit(x)
```

```python
plt.scatter(x[:,0],x[:,1],c=clustering.labels_, marker='o')
plt.show()
```

```python
clustering = AgglomerativeClustering(n_clusters=4,linkage='complete',affinity='manhattan')
```

```python
clustering.fit(x)
```

```python
plt.scatter(x[:,0],x[:,1],c=clustering.labels_, marker='o')
plt.show()
```
